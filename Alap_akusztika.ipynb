# A beszéd

A beszéd egy folytonos idejű jelekre, sztochasztikus, nem determinisztikus, kvázi periodikus jel.

**Sztochasztikus**: amelyeknél a múltbeli viselkedésből és a jelenkori értékükből nem lehet
következtetni a jövőbeli viselkedésre sztochasztikus jeleknek nevezzük. A sztochasztikus
jelek csak részben kezelhetőek matematikai módszerekkel. Leírásukhoz statisztikai
jellemzőket (várható érték-idő függvény, négyzetes középérték - idő függvény, variancia, stb.)
használunk.

**Determinisztikus jel**: a determinisztikus jelek matematikai összefüggések használatával
kezelhetők, leírhatók. Ahol a múltbeli viselkedésből és a jelenkori értékükből meg tudjuk
határozni, hogy a jövőben a jel miként viselkedik.

## Az akusztikai decibel definíciója

„A **decibel (dB)** két mennyiség arányának logaritmikus mértéke, amit széles körben használnak az akusztika, a fizika és az elektronika területén.”

„A decibel mértékegysége és dimenziója egy, hasonlóan a százalékhoz.”

„Alkalmazásának több előnye van: szélsőségesen nagy és kicsi értékek összehasonlítását teszi lehetővé, egyszerű összeadásra és kivonásra egyszerűsíti le az arányokkal való műveleteket, valamint a decibel logaritmikus skálája megfelel az emberi halló- és látószerv működésének.”

Általánosan a decibel (dB) két mennyiség arányának logaritmikus mértéke:

$X_{dB} = 10 log \frac{X}{X_0}$

Két tisztázásra váró fogalom: **hangintenzitás és hangnyomás**.

**Hangnyomás**: A hang terjedésekor a részecskék sűrűsödése és ritkulása egy adott pontban p(t)hang nyomásváltozást eredményez.  Az időfüggvénnyel rendkívül nehéz dolgozni, inkább valamilyen időbeli átlagot számolunk, leggyakrabban a négyzetes középértéket, az úgynevezett effektív értéket:

$p_{eff} = \sqrt{p^2(t)} = \sqrt{\frac{1}{t_1 - t_2} \int_{t_1}^{t_2} p^2(t)\mathrm{d}t}$

Mértékegysége: $Pa (\frac{N}{m^2})$.

**Hangintenzitás**: Az egységnyi az egységnyi felületre eső hangteljesítmény, amely egy adott hang hangosságát meghatározó fizikai paraméter. 

Mértékegysége $\frac{W}{m^2}$.

**Hangnyomás szint és hangintenzitás szint**:

$L_{I} = 10 * lg \frac{I}{I_0}, \quad I_0 = 10^-12 \frac{W}{m^2}$ <br /> <br />
$L_{p} = 20 * lg \frac{p}{p_0}, \quad p_0 = 2*10^-5 \frac{N}{m^2}$

## Mi az a dB SPL?

Jól halló alanyokon végzett kísérletek átlagából származik, hogy a hallásküszöb - az a minimális hangnyomásszint, amely már hangérzetet tud  kelteni - 20 μPa. A hangnyomás értékét  a normál légköri nyomástól való eltérés számértéke adja. Ezt nevezték el a 0 dB hangnyomásszintnek. Ha ehhez képest adjuk meg a hangnyomást, akkor szokás a dB SPL (SPL = Sound Pressure Level) kifejezést használni – amely egyébként megfelel az akusztikai decibel számértékének. Köztudott, hogy az ember a különböző frekvenciájú hangokat ugyanakkora hangnyomás ellenére különböző hangosságúnak érzékel.

## Autokorrelációs függvény és a HNR

Egy stacionárius $x(t)$ időfüggvény, $\tau$  laggal (eltolással) rendelkező autokorrelációs függvénye a következőképpen adható meg: 

$r_x(\tau) \equiv \int x(t) \cdot x(t+\tau) dt$  <br />

Az autokorrelációs függvénynek globális maximuma van $\tau = 0$-ban. Ha a függvénynek nullán kívüli globális maximuma is van, akkor a függvény periódikusnak tekinthető és létezik egy $T_0$ lagg érték, amit periódusnak nevezünk. Az összes globális maximum a periódusidő egészszámú többszörösénél helyezkedik el, azaz $nT_0$ pontokban, minden egészszámú $n$  értékre, úgy, hogy $r_x(nT_0) = r_x(0)$. Az alapfrekvenciát így a következőképpen definiáljuk:  <br />

$F_0 = \frac{1}{T0}$

Ha nincs globális maximuma az autokorrelációs függvénynek 0 pontban, még mindig létezhet lokális maximum. Ha a legnagyobb lokális maximum $\tau_{max}$ laggértéknél helyezkedik el és az $r_x\tau_{max}$ értéke nem elhanyagolható, akkor azt mondhatjuk, hogy a jelünk rendelkezik periodikus résszel. Ennek a résznek a harmonikus erőssége (harmonic strength) $R_0$, 0 és 1 közötti érték, ami egyenlő a normalizált autokorreláció lokális maximumával $r_x'(\tau_{max})$. 

$r_x'(\tau) \equiv \frac {r_x(\tau)}{r_x(0)}$

Ilyen $x(t)$ jelet tudunk készíteni ha veszünk egy $T_0$ periódusidejű $H(t)$ jelet és $N(t)$ zajt adunk hozzá. Arra következtethetünk az autokorrelációs függvényből, hogy ez a két jel nem korrelál egymással, a teljes függvény autokorrelációja a két rész autokorrelációjának az összege. <br />
Nulla lagg esetén tehát $r_x'(0) = r_H(0) + r_N(0)$.

<br />
Mivel egy jel autokorrelációja 0 lagg esetén egyenlő a jel teljesítményével, $\tau_max$ helyen normalizált függvény jelenti a periodikus (vagy harmonikus) összetevő relatív teljesítményét, valamint a kiegészítése jelenti a zajösszetevő relatív teljesítményét: <br />

$r_x'(\tau_{max}) \equiv \frac {r_H(\tau)}{r_x(0)}$ <br /> <br />
$1 - r_x'(\tau_{max}) \equiv \frac {r_N(\tau)}{r_x(0)}$ 


Ez lehetőséget ad nekünk, hogy definiáljuk a logaritmikus **HNR (harmonics-to-noise ratio)** arányt:

$HNR = 10 \cdot lg \frac {r_x'(\tau_{max})}{1 - r_x'(\tau_{max})}$ <br />

A HNR érték tökéletesen periodikus jelek esetében végtelen. A 0 dB HNR érték azt jelenti, hogy egyenlő energiamennyiség oszlik el a harmonikus és a zaj összetevők között.

```python
import parselmouth

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"]=8,8

plt.rcParams["font.family"] = "sans-serif"
plt.rcParams.update({'font.size': 12})

sns.set() # Use seaborn's default style to make attractive graphs

# Plot nice figures using Python's "standard" matplotlib library
snd = parselmouth.Sound("e.wav")
plt.figure()
plt.plot(snd.xs(), snd.values.T)
plt.xlim([snd.xmin, snd.xmax])
plt.title("Female [E] phone from continuous speech.")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
from parselmouth.praat import call

voiceID = "e.wav"
sound = parselmouth.Sound(voiceID) # read the sound

duration = call(sound, "Get total duration") # duration
print("Duration of sound: {} s".format(duration))
```

```python
def measurePitch(voiceID, f0min, f0max, unit):
    sound = parselmouth.Sound(voiceID) # read the sound
    pitch = call(sound, "To Pitch", 0.0, f0min, f0max) #create a praat pitch object
    meanF0 = call(pitch, "Get mean", 0, 0, unit) # get mean pitch
    stdevF0 = call(pitch, "Get standard deviation", 0 ,0, unit) # get standard deviation
    harmonicity = call(sound, "To Harmonicity (cc)", 0.01, 75, 0.1, 1.0)
    hnr = call(harmonicity, "Get mean", 0, 0)
    return meanF0, stdevF0, hnr


(meanF0, stdevF0, hnr) = measurePitch(voiceID, 50, 500, "Hertz")

print("Mean F0 value is: {} Hz".format(meanF0))
print("HNR value is: {}".format(hnr))
```

## Jitter és Shimmer

A jitter és a shimmer olyan akusztikai paraméterek amelyek a beszédjel alapfrekvencia és a hullámforma amplitúdó ciklikus változását számszerűsítik. Mindkét paramétert széles körben használják a kóros hangok azonosítására.
Jitter definíció szerint az alaphangidő periódusonkénti változékonysága(‘period-to-period variability of the pitch period).

Tehát, a beszédhangon belüli alapfrekvencia-ingadozást a jitterrel határozhatjuk meg. Az összefüggésben az egymást követő alapperiódusok időtartamainak egymáshoz való eltérését számoljuk.

### Jitter (ddp) 

$Jitter_{ddp} = \frac{\sum_{i=2}^{N-1} |2 \cdot T_i - T_{i-1} - T_{i+1}|}{\sum_{i=2}^{N-1} T_i} \cdot 100 [\%]$

### Shimmer (dda)

$Shimmer_{ddp} = \frac{\sum_{i=2}^{N-1} |2 \cdot A_i - A_{i-1} - A_{i+1}|}{\sum_{i=2}^{N-1} A_i} \cdot 100 [\%]$

```python
def measureJitterShimmer(voiceID, f0min, f0max, unit):
    sound = parselmouth.Sound(voiceID) # read the sound
    pointProcess = call(sound, "To PointProcess (periodic, cc)", f0min, f0max)
    ddpJitter = call(pointProcess, "Get jitter (ddp)", 0, 0, 0.0001, 0.02, 1.3)
    ddaShimmer = call([sound, pointProcess], "Get shimmer (dda)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    return ddpJitter, ddaShimmer

(ddpJitter, ddaShimmer) = measureJitterShimmer(voiceID, 50, 500, "Hertz")

print("Jitter ddp value is: {}".format(ddpJitter))
print("Shimmer dda value is: {}".format(ddaShimmer))
```
